Paso 4: Entrenamiento y optimización

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

# Split
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

best_models = {}

# KNN
knn_params = {'n_neighbors':[3,5,7], 'weights':['uniform','distance']}
knn = GridSearchCV(KNeighborsClassifier(), knn_params, cv=5, n_jobs=-1)
knn.fit(X_train, y_train)
best_models['KNN'] = knn.best_estimator_

# Gaussian NB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
best_models['GaussianNB'] = gnb

# Logistic Regression
lr_params = {'C':[0.1,1,10]}
lr = GridSearchCV(LogisticRegression(max_iter=1000), lr_params, cv=5, n_jobs=-1)
lr.fit(X_train, y_train)
best_models['LogisticRegression'] = lr.best_estimator_

# Decision Tree
dt_params = {'max_depth':[10,20,None], 'criterion':['gini','entropy']}
dt = GridSearchCV(DecisionTreeClassifier(), dt_params, cv=5, n_jobs=-1)
dt.fit(X_train, y_train)
best_models['DecisionTree'] = dt.best_estimator_

# SVM
svm_params = {'C':[1,10], 'kernel':['rbf'], 'gamma':['scale']}
svm = GridSearchCV(SVC(probability=True), svm_params, cv=5, n_jobs=-1)
svm.fit(X_train, y_train)
best_models['SVM'] = svm.best_estimator_

# Random Forest
rf_params = {'n_estimators':[100,200], 'max_depth':[None,20]}
rf = GridSearchCV(RandomForestClassifier(), rf_params, cv=5, n_jobs=-1)
rf.fit(X_train, y_train)
best_models['RandomForest'] = rf.best_estimator_

# Ver mejores parámetros
for name, model in best_models.items():
    print(name, model)
```

Paso 6: Función de preparación final

```python
def load_and_prepare(train_path, test_path):
    import pandas as pd

    drop_cols = ['LONGITUDE','LATITUDE','SPACEID','RELATIVEPOSITION',
                 'USERID','PHONEID','TIMESTAMP']

    # Cargar
    train = pd.read_csv(train_path)
    test = pd.read_csv(test_path)

    # Limpiar
    train = train.drop(columns=drop_cols)
    test = test.drop(columns=drop_cols)

    # Separar
    X_train = train.drop(columns=['FLOOR']).replace(100, -100)
    y_train = train['FLOOR']

    X_test = test.drop(columns=['FLOOR']).replace(100, -100)
    y_test = test['FLOOR']

    return X_train, X_test, y_train, y_test

# Uso
X_train, X_test, y_train, y_test = load_and_prepare(
    '/mnt/data/trainingData.csv',
    '/mnt/data/validationData.csv'
)
```

Paso 7: Evaluación

```python
import time
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize

results = []

for name, model in best_models.items():
    # Entrenamiento
    start = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start

    # Predicción
    start = time.time()
    y_pred = model.predict(X_test)
    test_time = time.time() - start

    # Métricas
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='macro')
    rec = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')

    # AUC
    try:
        y_prob = model.predict_proba(X_test)
        y_bin = label_binarize(y_test, classes=y.unique())
        auc = roc_auc_score(y_bin, y_prob, average='macro', multi_class='ovr')
    except:
        auc = None

    results.append([name, acc, prec, rec, f1, auc, train_time, test_time])

# Mostrar resultados
results_df = pd.DataFrame(results, columns=[
    'Modelo','Accuracy','Precision','Recall','F1','AUC','Train_time','Test_time'
])

print(results_df)
```
