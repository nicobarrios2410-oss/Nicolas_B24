```python
import pandas as pd

df = pd.read_csv('balanceSheetHistory_annually.csv')
df.head()
```

Paso 2 — Seleccionar columnas relevantes

```python
# Selección
df = df[['stock', 'endDate', 'cash']]

# Verificar
df.head()
```

Paso 3 — Separar por empresas

```python
# Convertir fecha
df['endDate'] = pd.to_datetime(df['endDate'])

# Ordenar
df = df.sort_values(by=['stock', 'endDate'])

# Eliminar nulos
df = df.dropna(subset=['cash'])

# Ver empresas con más datos
counts = df['stock'].value_counts()

# Elegir 3 empresas
stocks = counts.head(3).index.tolist()

print("Empresas seleccionadas:", stocks)

# Subsets
df_selected = df[df['stock'].isin(stocks)]
```

Paso 4 — Gráficas

```python
import matplotlib.pyplot as plt

for stock in stocks:
    data = df_selected[df_selected['stock'] == stock]

    plt.figure()
    plt.plot(data['endDate'], data['cash'])
    plt.title(f'Cash vs Time - {stock}')
    plt.xlabel('Fecha')
    plt.ylabel('Cash')
    plt.xticks(rotation=45)
    plt.grid()
    plt.show()
```

Paso 5 — División 80/20

```python
train_data = {}
test_data = {}

for stock in stocks:
    data = df_selected[df_selected['stock'] == stock]

    split = int(len(data) * 0.8)

    train_data[stock] = data.iloc[:split]
    test_data[stock] = data.iloc[split:]
```

Paso 6 — Modelo de regresión lineal

```python
from sklearn.linear_model import LinearRegression

models = {}
predictions = {}

for stock in stocks:
    train = train_data[stock].copy()
    test = test_data[stock].copy()

    # Crear lag en train
    train['lag1'] = train['cash'].shift(1)
    train = train.dropna()

    # Verificar que haya datos suficientes
    if len(train) < 2:
        print(f"{stock}: No hay suficientes datos para entrenar")
        continue

    X_train = train[['lag1']]
    y_train = train['cash']

    model = LinearRegression()
    model.fit(X_train, y_train)

    # Crear lag en test
    test['lag1'] = test['cash'].shift(1)
    test = test.dropna()

    # 🔴 VALIDACIÓN CLAVE
    if test.empty:
        print(f"{stock}: No hay suficientes datos en test para predecir")
        continue

    X_test = test[['lag1']]

    y_pred = model.predict(X_test)

    models[stock] = model
    predictions[stock] = (test, y_pred)
```

Paso 7 — Gráfica real vs predicho

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

metrics = {}

for stock in stocks:

    # 🔴 VALIDAR SI EXISTE EN PREDICCIONES
    if stock not in predictions:
        print(f"{stock}: No tiene predicciones, se omite")
        continue

    test, y_pred = predictions[stock]

    # 🔴 VALIDAR QUE NO ESTÉ VACÍO
    if test.empty or len(y_pred) == 0:
        print(f"{stock}: Datos insuficientes para graficar")
        continue

    # 📊 GRÁFICA
    plt.figure()
    plt.plot(test['endDate'], test['cash'], label='Real')
    plt.plot(test['endDate'], y_pred, linestyle='--', label='Predicho')

    plt.title(f'Real vs Predicho - {stock}')
    plt.xlabel('Fecha')
    plt.ylabel('Cash')
    plt.legend()
    plt.grid()
    plt.xticks(rotation=45)
    plt.show()

    # 📈 MÉTRICAS
    mse = mean_squared_error(test['cash'], y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(test['cash'], y_pred)

    metrics[stock] = (mse, rmse, r2)
```

Paso 8 — Métricas

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd

resultados = []

for stock in stocks:

    # Validar que exista predicción
    if stock not in predictions:
        print(f"{stock}: sin predicciones")
        continue

    test, y_pred = predictions[stock]

    if test.empty or len(y_pred) == 0:
        print(f"{stock}: datos insuficientes")
        continue

    y_true = test['cash']

    # Métricas
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)

    resultados.append({
        'Empresa': stock,
        'MSE': mse,
        'RMSE': rmse,
        'R2': r2
    })

# Convertir a tabla
tabla_resultados = pd.DataFrame(resultados)

# Mostrar tabla
print(tabla_resultados)
```

Pequeña conclusión:

El RMSE y MSE permiten evaluar el error en las mismas unidades de cash, donde valores más bajos indican mejor precisión del modelo.

El RMSE es más fácil de interpretar porque está en la misma escala que los datos originales.

El R² muestra qué proporción de la variabilidad del cash es explicada por el modelo; valores cercanos a 1 indican buen ajuste, mientras que valores cercanos a 0 o negativos indican bajo desempeño.

Si alguna empresa presenta un RMSE alto o R² bajo, esto sugiere que el modelo no está capturando bien el comportamiento de la serie temporal.

Para mejorar el modelo se podrían incluir más variables (como otros indicadores financieros), utilizar más rezagos (lag2, lag3), aplicar transformaciones o emplear modelos más avanzados como ARIMA o Random Forest.

Preguntas de Análisis
1. Variables usadas

Se utilizó como variable principal un rezago (lag1) de cash, es decir, el valor del periodo anterior.
Es relevante porque en series temporales los valores pasados influyen directamente en los futuros.

2. División 80/20

Se dividieron los datos por empresa respetando el tiempo:

80% inicial → entrenamiento
20% final → prueba

No se usó shuffle porque rompería el orden temporal y generaría predicciones irreales.

3. MSE vs RMSE
MSE: error promedio al cuadrado (penaliza errores grandes).
RMSE: raíz del MSE, en la misma unidad que cash.

El RMSE es más fácil de interpretar.

4. Interpretación de R²

El R² indica qué tanto el modelo explica la variación del cash.
Valores cercanos a 1 son buenos, pero en series temporales puede ser engañoso si hay tendencias.

5. Cómo mejorar el modelo

Si el modelo es malo:

Añadir más rezagos (lag2, lag3)
Incluir más variables financieras
Aplicar transformaciones (log)
Usar modelos más avanzados (ARIMA, Random Forest, LSTM)
